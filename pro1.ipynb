```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
cities={'A':'SEACET','B':'K R PURM','C':'ITI','D':'TIN FACTORY','E':'RM NAGER'}
roads=[('A','B'),('A','C'),('B','D'),('C','E'),('D','E')]
```

```python
G.add_edges_from(roads)
G.add_nodes_from(cities.keys())
```

```python
pos=nx.spring_layout(G)
nx.draw(G,pos, labels=cities,node_size=2000)
plt.title('MAP')
plt.show()
```

```python
def bfs(graph, start):
    visited=set()
    queue=[start]
    bfs_res=[]
    while queue:
        node=queue.pop(0)
        if node not in visited:
            bfs_res.append(node)
            visited.add(node)
            queue.extend(neighbor for neighbor in graph.neighbors(node) if neighbor not in visited)
    return bfs_res
```

```python
bfs_res=bfs(G,'A')
```

```python
print('bfs_res',[cities[node] for node in bfs_res])
```

```python
def dfs(graph,start):
    visited=set()
    dfs_res=[]
    def dfs_rec(node):
        nonlocal visited
        visited.add(node)
        dfs_res.append(node)
        for neighbor in graph.neighbors(node):
            if neighbor not in visited:
                dfs_rec(neighbor)
    dfs_rec(start)
    return dfs_res
```

```python
Start_city='A'
bfs_res=bfs(G,Start_city)
dfs_res=dfs(G,Start_city)
print("BFS Result",[cities[node] for node in bfs_res])
print("DFS Result",[cities[node] for node in dfs_res])
```
