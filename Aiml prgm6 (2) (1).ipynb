```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB

# Load the dataset
data = pd.read_csv(r'C:\Users\nu\Desktop\data_Sets\PlayTennis.csv')
print("The first 5 values of data are:\n", data.head())

# Obtain train data and train output
X = data.iloc[:, :-1]
print("\nThe First 5 values of train data are:\n", X.head())

y = data.iloc[:, -1]
print("\nThe first 5 values of Train output are:\n", y.head())

# Initialize LabelEncoder
le_outlook = LabelEncoder()
le_temperature = LabelEncoder()
le_humidity = LabelEncoder()

# Encode categorical features in X
X.loc[:, 'Outlook'] = le_outlook.fit_transform(X['Outlook'])
X.loc[:, 'Temperature'] = le_temperature.fit_transform(X['Temperature'])
X.loc[:, 'Humidity'] = le_humidity.fit_transform(X['Humidity'])

print("\nNow the Train data is:\n", X.head())

# Encode target variable y
le_PlayTennis = LabelEncoder()
y = le_PlayTennis.fit_transform(y)
print("\nNow the training output is:\n", y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Initialize Gaussian Naive Bayes classifier
classifier = GaussianNB()

# Fit the classifier to the training data
classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = classifier.predict(X_test)

# Calculate and print the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB

# Load the dataset
data = pd.read_csv(r'C:\Users\nu\Desktop\data_Sets\PlayTennis.csv')
print("The first 5 values of data are:\n", data.head())

# Obtain train data and train output
X = data.iloc[:, :-1]
print("\nThe First 5 values of train data are:\n", X.head())

y = data.iloc[:, -1]
print("\nThe first 5 values of Train output are:\n", y.head())

# Initialize LabelEncoder for categorical columns
le_outlook = le_temperature =le_humidity =le_wind = LabelEncoder()

# Encode categorical features
X.loc[:, 'Outlook'] = le_outlook.fit_transform(X['Outlook'])
X.loc[:, 'Temperature'] = le_temperature.fit_transform(X['Temperature'])
X.loc[:, 'Humidity'] = le_humidity.fit_transform(X['Humidity'])
X.loc[:, 'Wind'] = le_wind.fit_transform(X['Wind'])

print("\nNow the Train data is:\n", X.head())


le_PlayTennis = LabelEncoder()
y = le_PlayTennis.fit_transform(y)
print("\nNow the training output is:\n", y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Initialize Gaussian Naive Bayes classifier
classifier = GaussianNB()

# Fit the classifier to the training data
classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = classifier.predict(X_test)

# Calculate and print the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
